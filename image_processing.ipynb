```python
# Display RGB Channels of the image
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(image_1[:,:,0], cmap='Reds')
axs[1].imshow(image_1[:,:,1], cmap='Greens')
axs[2].imshow(image_1[:,:,2], cmap='Blues')
axs[0].set_title('Reds')
axs[1].set_title('Greens')
axs[2].set_title('Blues')
```

Preparing Grey Image

```python
img_gray = cv2.cvtColor(image_1, cv2.COLOR_RGB2GRAY)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_gray, cmap='Greys')
ax.axis('off')
ax.set_title('Grey Image')
plt.show()
```

Resizing the image

```python
img_resized = cv2.resize(image_1, None, fx=0.2, fy=0.2)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resized)
ax.axis('off')
plt.show()
```

```python
img_resize = cv2.resize(image_1, (200, 100))
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resize)
ax.axis('off')
plt.show()
```

```python
img_resize = cv2.resize(image_1, (5000, 5000), interpolation = cv2.INTER_AREA)
fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(img_resize)
ax.axis('off')
plt.show()
```

Edge Detection

```python
#edge detection
kernel_edge_detection = np.array([[-1,-1,-1],
                                  [-1,8,-1],
                                   [-1,-1,-1]])

edged = cv2.filter2D(image_1, -1, kernel_edge_detection)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(edged)
ax.axis('off')
ax.set_title('Edge Detection')
plt.show()
```

```python
#blurring the image
kernel_blurring = (np.ones((3,3)))/20
kernel_blurring
blurred = cv2.filter2D(image_1, -1, kernel_blurring)

fig, ax = plt.subplots(figsize=(8, 8))
ax.imshow(blurred)
ax.axis('off')
ax.set_title('Blurred Image')
plt.show()
```

```python
cv2.imwrite('cat_image_blur.png',blurred)
```
